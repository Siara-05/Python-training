```python
def sub():
    a = 55
    b = 98
    c = b-a
    print("subraction of two numbers is: ",c )

sub()
add()
```

## Function with parameters

```python
def add(a,b):
    return a + b

result = add(12 , 9)
print("After addition of 2 numbers, the resukt is:",result)
```

### Subtraction using function with parameters

```python
def sub(a,b,c):
    return a - b - c 

result = sub (98, 56, 2)
print("After subtraction of 3 numbers, the result is: " ,result)
```
