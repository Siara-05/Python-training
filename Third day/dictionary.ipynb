## $Dictionary$
- Created using { }
- It is in a key : value psir
- Muttable
- unordered collection of elements 

```python
a = { "name": "Subha", "Age" : 21}
print(a)
print(type(a))
```

### ▪️Methods of dictionary

#### Accessing the values

```python
print(a['name'])
print(a['Age'])
```

### Adding or updating items 

```python
a["Prem"]=12
a

a['name']= "Sia"
a
```

Removing the elements from the dict

```python
a.__delitem__('name')
print(a)
```

```python
a = {'Product': 'Diary', 'sell':10, 'due-date':'28 feb'}
print(a)

a = {['names':'Black dairy','White','gray','colourfull']}
print(a)
```
