# Sets
- Unordered collection of unique elements 
- Denoted by {}
- Mutable



```python
a = [1,2,3,44,44,44,55,55,99,5,6]     # List
print(a)

b = (1,2,3,44,44,44,55,55,99,5,6)     # Tuple
print(b)

c = {1,2,3,44,44,44,55,55,99,5,6, "Python" , "Javascript"}     # Set
print(c)

```

## Set operations 

```python
a = {1,2,3,4,5}
print(type(a))
print(a)

b = {"python":"backend",
     "javascript":"frontend"}
print(type(b))
print(b)
```

### Discard Function

- Remove function gives an error if an element doesn't exist in the list
- Discard function removes the element if exist, and if doesn't exist does nothing (no error)
- Use discard when we dont't want error
- Use remove if we want error

```python
a.remove(5)
print(a)

a.discard(4)
print(a)

a.discard(4)
print(a)
```
