### ***List Operations***

```python
list_num = [26, 13, "prem" , "subha", 'suki' , 2024 ]

print ( list_num [0])
print ( list_num [::-1])
print ( list_num [0:4])
print ( list_num [1:4])
print ( list_num [:4])
print ( list_num [:5])
```

## Write a program to accapte a value from user and specify its even or odd

```python
def even_odd():
    num = int(input(" Enter your number : "))
    

    if num > 0:

     if num % 2 ==0 :
        print(f'The given number {num} is even ')

     else:
        print(f'The given number {num} is odd ')

even_odd()


```
